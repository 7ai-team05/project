```python
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

# 이미지 불러오기
img = Image.open('data\IMG_9312.JPG')

# 랜덤으로 224x224 영역을 자름
transform = transforms.RandomCrop((224, 224))

cropped_img = transform(img)

# 출력
plt.imshow(cropped_img)
plt.title("Random Cropped")
plt.axis('off')
plt.show()
```

```python
from PIL import Image
from torchvision import transforms
import os

# 원본 이미지 폴더 & 저장 폴더 설정
input_dir = "C:/Users/USER/Downloads/image_data/1. clean"
output_dir = "C:/Users/USER/Desktop/aug_clean"
os.makedirs(output_dir, exist_ok=True)

# 증강 파이프라인 정의
transform = transforms.Compose([
    transforms.RandomResizedCrop(224),         # 랜덤 크롭
    transforms.RandomHorizontalFlip(),         # 좌우 반전
    transforms.RandomRotation(15),             # 회전
    transforms.ColorJitter(0.2, 0.2, 0.2),      # 밝기/대비/채도 변화
])

# 증강 이미지 생성
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(input_dir, filename)
        img = Image.open(img_path)

        for i in range(5):  # 한 이미지당 5개씩 증강
            augmented_img = transform(img)
            output_path = os.path.join(output_dir, f"{filename[:-4]}_aug{i}.jpg")
            augmented_img.save(output_path)

print("✅ 데이터 증강 및 저장 완료")
```
