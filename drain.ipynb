```python
import io
import requests
from PIL import Image
import gradio as gr

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 1) ì•„ë˜ ë‘ ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê°’ìœ¼ë¡œ ìˆ˜ì •í•˜ì„¸ìš”: (ë¹—ë¬¼ë°›ì´ ì—¬ë¶€ íŒë³„ ëª¨ë¸)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
DETECTION_KEY = "5sloqMYWiZSpfCd7HZgQ9ZpfuQAXnRWwjSw648WjXGr5Fy5f7imcJQQJ99BFACYeBjFXJ3w3AAAIACOGBDCR"
DETECTION_ENDPOINT_URL = "https://7aiteam05ai012-prediction.cognitiveservices.azure.com/customvision/v3.0/Prediction/f360eeb2-f8df-441e-8bed-f27d3ef1797a/detect/iterations/Iteration2/image"

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 2) ì•„ë˜ ë‘ ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê°’ìœ¼ë¡œ ìˆ˜ì •í•˜ì„¸ìš”: (ì˜¤ì—¼ë„ ë¶„ë¥˜ ëª¨ë¸: Clean vs Heavy)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
POLLUTION_KEY = "BBvYKDdr5RDpSMjG34Z2XXw3hLxzlAQkktCPXwHTLleSagQPHGg0JQQJ99BEACYeBjFXJ3w3AAAIACOGH9bC"
POLLUTION_ENDPOINT_URL = "https://7aiteam05cv-prediction.cognitiveservices.azure.com/customvision/v3.0/Prediction/2ae6121f-4235-4dce-bf2a-fbace9444880/classify/iterations/Iteration12/image"


def predict_sequence(img: Image.Image) -> tuple[dict, dict, str]:
    """
    1) Gradioë¡œ ì…ë ¥ë°›ì€ PIL ì´ë¯¸ì§€ë¥¼ JPEG ë°”ì´íŠ¸ë¡œ ë³€í™˜
    2) [ë¹—ë¬¼ë°›ì´ ì—¬ë¶€ ëª¨ë¸]ì— POST ìš”ì²­ -> ê²°ê³¼ë¥¼ { 'drain':í™•ë¥ , 'not_drain':í™•ë¥  }ë¡œ ì •ë¦¬
    3) 'drain' í™•ë¥ ì´ 'not_drain'ë³´ë‹¤ ë†’ìœ¼ë©´(=ë¹—ë¬¼ë°›ì´ íŒì •):
         -> [ì˜¤ì—¼ë„ ëª¨ë¸]ì— POST ìš”ì²­ -> ê²°ê³¼ë¥¼ { 'clean':í™•ë¥ , 'heavy':í™•ë¥  }ë¡œ ì •ë¦¬
       ì•„ë‹ˆë©´(=ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹ˆë©´) ë‘ ë²ˆì§¸ ëª¨ë¸ í˜¸ì¶œ ìƒëµ
    4) ìƒíƒœ ë©”ì‹œì§€ë¥¼ ì„¸ ê°€ì§€ ê²½ìš°ë¡œ ë¶„ê¸°í•˜ì—¬ ë¬¸ìì—´ë¡œ ìƒì„±
       - ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹˜
       - ì˜¤ì—¼ë„(heavy) < 70%
       - ì˜¤ì—¼ë„(heavy) â‰¥ 70% & < 90%
       - ì˜¤ì—¼ë„(heavy) â‰¥ 90%
    5) Gradioì— (detection_dict, pollution_dict, status_text) íŠœí”Œë¡œ ë°˜í™˜
    """
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # A) PIL Image â†’ JPEG ë°”ì´íŠ¸ ë³€í™˜ (ê³µí†µ)
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    with io.BytesIO() as buffer:
        img.save(buffer, format="JPEG")
        image_bytes = buffer.getvalue()

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # B) ë¹—ë¬¼ë°›ì´ ì—¬ë¶€ ëª¨ë¸ í˜¸ì¶œ (Detection)
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    headers_det = {
        "Prediction-Key": DETECTION_KEY,
        "Content-Type": "application/octet-stream"
    }
    response_det = requests.post(
        DETECTION_ENDPOINT_URL,
        headers=headers_det,
        data=image_bytes
    )
    response_det.raise_for_status()
    result_det = response_det.json()
    preds_det = result_det.get("predictions", [])

    # ì¶œë ¥ ë”•ì…”ë„ˆë¦¬ë¥¼ ëª¨ë‘ ì†Œë¬¸ì í‚¤ë¡œ ì •ë¦¬
    detection_dict = {}
    for pred in preds_det:
        name = pred.get("tagName", "")
        prob = pred.get("probability", 0.0)
        detection_dict[name.lower()] = prob
    # ì˜ˆ) detection_dict == { 'drain': 0.85, 'not_drain': 0.15 }

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # C) ë¹—ë¬¼ë°›ì´ íŒì • ê¸°ì¤€: drain í™•ë¥  vs not_drain í™•ë¥ 
    #    (ì—¬ê¸°ì„œëŠ” ë” ë†’ì€ ìª½ì„ ì„ íƒ)
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    drain_prob = detection_dict.get("street_drain", 0.0)
    not_drain_prob = detection_dict.get("unstreet_drain", 0.0)

    # ê¸°ë³¸ê°’: ì˜¤ì—¼ë„ ëª¨ë¸ í˜¸ì¶œì„ ê±´ë„ˆë›¸ ë•Œ ì‚¬ìš©
    pollution_dict = { "clean": 0.0, "heavy": 0.0 }
    status_text = ""

    if drain_prob > not_drain_prob:
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # D) ë¹—ë¬¼ë°›ì´(Drain)ìœ¼ë¡œ íŒì •ë˜ì—ˆìœ¼ë¯€ë¡œ ì˜¤ì—¼ë„(í´ë¦°/í—¤ë¹„) ëª¨ë¸ í˜¸ì¶œ
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        headers_pol = {
            "Prediction-Key": POLLUTION_KEY,
            "Content-Type": "application/octet-stream"
        }
        response_pol = requests.post(
            POLLUTION_ENDPOINT_URL,
            headers=headers_pol,
            data=image_bytes
        )
        response_pol.raise_for_status()
        result_pol = response_pol.json()
        preds_pol = result_pol.get("predictions", [])

        # clean/heavy í™•ë¥ ì„ ì†Œë¬¸ì í‚¤ë¡œ ì •ë¦¬
        pollution_dict = {}
        for pred in preds_pol:
            name = pred.get("tagName", "")
            prob = pred.get("probability", 0.0)
            pollution_dict[name.lower()] = prob
        # ì˜ˆ) pollution_dict == { 'clean': 0.10, 'heavy': 0.90 }

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # E) pollution_dict['heavy'] ê°’ì— ë”°ë¼ ìƒíƒœ ë©”ì‹œì§€ ë¶„ê¸°
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        heavy_prob = pollution_dict.get("heavy", 0.0)
        if heavy_prob >= 0.9:
            status_text = "ğŸš¨ Heavy ìˆ˜ì¤€: í™•ë¥  â‰¥ 90% (ë§¤ìš° ë§‰í˜)"
        elif heavy_prob >= 0.7:
            status_text = "âš ï¸ Heavy ìˆ˜ì¤€: í™•ë¥  â‰¥ 70% (ì¤‘ê°„/ë†’ìŒ ë§‰í˜)"
        elif heavy_prob >= 0.6:
            status_text = "âœ… Heavy í™•ë¥  â‰¥ 60% (í¬ê²Œ ë§‰íŒ ìƒíƒœ ì•„ë‹˜)"
        else:
            status_text = "âœ… Clean"
    else:
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # F) ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹ ë•Œ: pollution_dictëŠ” 0.0 ê¸°ë³¸ê°’ ì‚¬ìš©, ë©”ì‹œì§€ ì„¤ì •
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        status_text = "âŒ ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹™ë‹ˆë‹¤."

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # G) Gradioë¡œ ë°˜í™˜: (ë¹—ë¬¼ë°›ì´ íŒë³„ dict, ì˜¤ì—¼ë„ dict, ìƒíƒœ ë©”ì‹œì§€)
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    return detection_dict, pollution_dict, status_text


if __name__ == "__main__":
    demo = gr.Interface(
        fn=predict_sequence,
        inputs=gr.Image(type="pil"),
        outputs=[
            gr.Label(label="ë¹—ë¬¼ë°›ì´ ì—¬ë¶€ (Drain vs NotDrain)"),
            gr.Label(label="ì˜¤ì—¼ë„ (Clean vs Heavy)"),
            gr.Text(label="ìƒíƒœ ë©”ì‹œì§€")
        ],
        title="Custom Vision: ë¹—ë¬¼ë°›ì´ ì—¬ë¶€ â” ì˜¤ì—¼ë„ ìˆœì°¨ ë¶„ë¥˜",
        description="""
        1. ë¨¼ì € ì—…ë¡œë“œëœ ì‚¬ì§„ì´ ë¹—ë¬¼ë°›ì´ì¸ì§€ ì•„ë‹Œì§€ Custom Vision ëª¨ë¸ë¡œ íŒì •í•©ë‹ˆë‹¤.  
        2. 'ë¹—ë¬¼ë°›ì´(drain)'ë¡œ íŒì •ë˜ë©´, ë‘ ë²ˆì§¸ ëª¨ë¸ì„ í˜¸ì¶œí•˜ì—¬ 'clean' vs 'heavy' ì˜¤ì—¼ë„ë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.  
        3. ê²°ê³¼ëŠ” ê°ê° í™•ë¥ ê³¼ ìƒíƒœ ë©”ì‹œì§€ë¡œ ë³´ì—¬ì¤ë‹ˆë‹¤.  
        - ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹Œ ê²½ìš°: ìƒíƒœ ë©”ì‹œì§€ì— 'âŒ ë¹—ë¬¼ë°›ì´ê°€ ì•„ë‹™ë‹ˆë‹¤.' ì¶œë ¥  
        - ë¹—ë¬¼ë°›ì´ì¸ ê²½ìš°, heavy í™•ë¥ ì— ë”°ë¼ 70%Â·90% ê²½ê³„ë¡œ ìƒíƒœ ë©”ì‹œì§€ ë¶„ê¸°  
        """,
        allow_flagging="never"
    )
    demo.launch(share=True)
```
