```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

# 1. CSV 파일 불러오기
df = pd.read_csv("C:/Users/USER/Downloads/빗물받이_customvision-20250601T162709Z-1-001/빗물받이_customvision/project/project/results/school_attack 1.csv")

# 2. tag_info 컬럼(문자열)을 dict로 변환
df['tag_info'] = df['tag_info'].apply(ast.literal_eval)

# 3. total_tag 값만 추출하여 새로운 칼럼 생성
df['total_tag'] = df['tag_info'].apply(lambda x: x.get('total_tag', 0))

# 4. 학교별 total_tag 합계 계산 및 내림차순 정렬
school_tag_sum = (
    df
    .groupby('school')['total_tag']
    .sum()
    .sort_values(ascending=False)
)

# 5. 막대그래프로 시각화
plt.figure(figsize=(10, 6))
plt.bar(school_tag_sum.index, school_tag_sum.values)
plt.xticks(rotation=45, ha='right')   # 레이블 45도 회전
plt.xlabel('학교명')
plt.ylabel('총 태그 개수 합계')
plt.title('학교별 총 태그 개수 순위')
plt.tight_layout()  # 레이아웃 자동 조정
plt.show()
```
