```python
from azure.cognitiveservices.vision.customvision.training import CustomVisionTrainingClient
from msrest.authentication import ApiKeyCredentials
import os
import requests

# === 설정 ===
ENDPOINT = "https://1dt-team3-aivision.cognitiveservices.azure.com/"
TRAINING_KEY = "1NJVieYcMfABbCFsniV7ZSKT8aBMwwNc5Xw16uPrhqpEajvLNjLDJQQJ99BEACYeBjFXJ3w3AAAEACOGBubo"
PROJECT_ID = "405bfcac-de54-482e-a5c0-e7da98a63606"
BASE_SAVE_DIR = "exported_images_test"
IMAGES_PER_PAGE = 100  # 한 번에 가져올 개수 (최대 256)

# 클라이언트 초기화
credentials = ApiKeyCredentials(in_headers={"Training-key": TRAINING_KEY})
trainer = CustomVisionTrainingClient(ENDPOINT, credentials)

# 태그 정보 가져오기
tags = trainer.get_tags(PROJECT_ID)
tag_dict = {tag.id: tag.name for tag in tags}

# 태그 폴더 생성
for tag_name in tag_dict.values():
    os.makedirs(os.path.join(BASE_SAVE_DIR, tag_name), exist_ok=True)

# 모든 이미지 가져오기 (페이지네이션)
skip = 0
all_images = []

while True:
    images = trainer.get_images(PROJECT_ID, take=IMAGES_PER_PAGE, skip=skip, tag_ids=None)
    if not images:
        break
    all_images.extend(images)
    skip += IMAGES_PER_PAGE

print(f"📦 총 이미지 수: {len(all_images)}")

# 이미지 다운로드 및 저장
for img in all_images:
    if img.original_image_uri:
        try:
            img_data = requests.get(img.original_image_uri).content
        except:
            print(f"❌ 다운로드 실패: {img.id}")
            continue

        # 태그별로 저장
        for tag in img.tags:
            tag_name = tag_dict[tag.tag_id]
            tag_folder = os.path.join(BASE_SAVE_DIR, tag_name)
            filename = os.path.join(tag_folder, f"{img.id}.jpg")

            with open(filename, 'wb') as f:
                f.write(img_data)

        print(f"✔ 저장: {img.id} → {[tag_dict[tag.tag_id] for tag in img.tags]}")
```
